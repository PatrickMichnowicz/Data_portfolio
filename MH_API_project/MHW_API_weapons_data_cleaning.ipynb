```python
# Import useful libraries
import numpy as np
import pandas as pd
import requests as req
```

```python
pd.set_option("display.max_row", 30)
pd.set_option("display.max_columns", None)
pd.set_option('display.max_colwidth', 50)
```

```python
# Import API from url

url1 = "https://mhw-db.com/weapons"
response = req.get(url1)
response
```

```python
data = pd.json_normalize(response.json())
data
```

```python
cols = data.columns.values
cols
```

```python
# Subset with information to be used
wdata = data[['id', 'name', 'type', 'damageType', 'attack.display', 'attack.raw', 'elements', 'attributes.affinity', 'durability', 'slots']]
wdata
```

```python
wdata.dtypes
```

```python
# 1 of the weapons has a double type and data is contained in a list containing 2 dictionaries. 
# Upon exploding them, it creates 2 entries, 1 for each elements, thus adding an additional row
wdata[wdata["id"]==343]["elements"].values
```

```python
# Solution: since this is only for 1 row and the "damage" value is the same for both elements,
# I replaced the value with the following: 
wdata.iloc[342,6] = [{'type': 'ice/blast', 'damage': 240, 'hidden': False}]
```

```python
# Unpacking values from "elements"
elements = pd.DataFrame(wdata.loc[:,["id","elements"]])
elements
```

```python
# Replace empty data as we cannot drop the empty lists -> we must set a default value for the empty lists 
# Why this method? Seems Pandas does not recognize empty list as NaN, nor as "[]". Hence, we need a reference column.
# 1. Create list with bool values
empty = []
for i in range(len(elements)):
    if len(elements["elements"][i]) == 0:
        empty.append(True) # elements is empty
    else:
        empty.append(False) # elements is filled in
```

```python
# 2. Add column "empty" to our DF
elements["empty"] = empty
elements
```

```python
# 3. Based on the values in "empty", replace the [] with a default value for rows that have a True value in "empty"
for i in range(len(elements)):
    if elements.iloc[i,2] == True:
        elements.iloc[i,1] = [{"type":"None", "damage":0, "hidden":False}]
    else:
        continue
```

```python
elements = elements.drop(columns="empty") # drop "empty" column (no longer needed)
elements = elements.explode("elements") # converts lists into dictionaries
```

```python
# for each row, return all values from the dictionnaries and append them to our DF as new columns 
def extract_values(dictionary):
    type = dictionary["type"]
    damage = dictionary["damage"]
    hidden = dictionary["hidden"]
    return type, damage, hidden

elements[["element_type","element_damage", "element_hidden"]] = elements["elements"].apply(lambda x:pd.Series(extract_values(x)))
```

```python
# As expected, 1 of the weapons has a new element-type, which is the combination of its original elements ('ice/blast')
elements["element_type"].unique()
```

```python
elements
```

```python
# Unpack durability
# Problem 1: 6 levels of durability (basic level + 1 level per level of Sharpness skill)
# Problem 2: 3 weapon types do not have durability 
# 1) Create a DF, excluding weapons without durability
durability = wdata[~wdata["type"].isin(['light-bowgun', 'heavy-bowgun', 'bow'])]
durability = durability.loc[:,["id","durability"]]
```

```python
# Convert all lists into dictionaries
durability_all = durability.explode("durability")
durability_all
```

```python
# 2) Get all values from every dictionary
def extract_values(dictionary):
    red = dictionary["red"]
    orange = dictionary["orange"]
    yellow = dictionary["yellow"]
    green = dictionary["green"]
    blue = dictionary["blue"]
    white =  dictionary["white"]
    purple = dictionary["purple"]
    return red, orange, yellow, green, blue, white, purple

durability_all[["red","orange", "yellow", "green", "blue", "white", "purple"]] = durability_all["durability"].apply(lambda x:pd.Series(extract_values(x)))
durability_all
```

```python
# We wish to get the base and max durability for all weapons (add total column)
durability_all["total_durability"] = (durability_all["red"]
                                  + durability_all["orange"]
                                  + durability_all["yellow"]
                                  + durability_all["green"]
                                  + durability_all["blue"]
                                  + durability_all["white"]
                                  + durability_all["purple"])
durability_all[durability_all["id"]==1]
```

```python
# Select min and max values for durability
min_durability = durability_all.iloc[::6]
max_durability = durability_all.iloc[5::6]
```

```python
durability["base_durability"] = min_durability["total_durability"]
durability["max_durability"] = max_durability["total_durability"]
```

```python
# To use in final weapons DF
durability
```

```python
# Save all durability values in another csv (for future use)
full_durability = durability_all.drop(columns=["durability", "total_durability"])
```

```python
# Insert a reference column to list all values for each weapon
full_durability.insert(1, "level_durability", None)
```

```python
# Add values through iloc slicing
full_durability.iloc[0::6,1]= "base"
full_durability.iloc[1::6,1]= "sharpness +1"
full_durability.iloc[2::6,1]= "sharpness +2"
full_durability.iloc[3::6,1]= "sharpness +3"
full_durability.iloc[4::6,1]= "sharpness +4"
full_durability.iloc[5::6,1]= "sharpness +5"
```

```python
# Unpack slots : get total amount of slots for each weapon
slots = pd.DataFrame(wdata.loc[:,["id","slots"]])
len(slots.iloc[700,1])
```

```python
total_slots = []
for i in range(len(slots)):
    total_slots.append(len(slots.iloc[i,1]))
```

```python
# To use
slots["total_slots"] = total_slots
slots
```

```python
# Finalize DF and export to workable CSV
wdata.loc[:,"total_slots"] = slots.loc[:,"total_slots"]
wdata.loc[:,"base_durability"] = durability.loc[:,"base_durability"]
wdata.loc[:,"max_durability"] = durability.loc[:,"max_durability"]
wdata.loc[:,["element_type", "element_damage", "element_hidden"]] = elements.loc[:,["element_type",	"element_damage", "element_hidden"]]
```

```python
# Get rid of cleaned columns
weapon_final = wdata.drop(columns=["elements","durability", "slots"])
```

```python
# Clean nulls, replace "True" and "False" with "Yes" and "No" (more user friendly in this case)
weapon_final.loc[:,"element_hidden"] = weapon_final["element_hidden"].replace(to_replace=[False, True], value=["No", "Yes"])
weapon_final.loc[:,"attributes.affinity"] = weapon_final["attributes.affinity"].fillna(0)
```

```python
# Explicit certain column names
weapon_to_export = weapon_final.rename(columns={
    "damageType" : "damage_type",
    "attack.display" : "attack",
    "attack.raw" : "raw",
    "attributes.affinity" : "affinity",
    "element_type" : "elements"
})
```
